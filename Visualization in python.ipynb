```python
# Works only on COLAB

def read_csv_from_github(url):
    import requests
    from io import StringIO
    response = requests.get(url)
    data = response.text
    return pd.read_csv(StringIO(data))


url = 'https://github.com/bgagl/ML_Individual_Differences/raw/5b70d36362172bb50d5be984e8c97526dda26bd2/datasets/data.csv'
df = read_csv_from_github(url=url)
df.head()
```

**Histogram**

A histogram is a graphical representation of the distribution of a continuous variable. The graph shows the frequency of different numerical ranges. The x-axis represents the values of the variable, and the y-axis represents the frequency of those values. The graph is divided into several "bins". The height of each bar in the histogram corresponds to the number of observations in the bin.

Here we use the `hist()` function and add labels and title. Note to select a variable in a data frame with `df["variable name"]`.


```python
df["Random Float"]
```

```python
plt.hist(df["Random Float"])
plt.xlabel('Random Float')
plt.ylabel('Frequency')
plt.title('My Histogram')
plt.show()
```

**Scatter Plot**

A scatter plot displays values of typically two variables of a data set to observe relationships between variables. The data is shown as a collection of points, each having the value of one variable determining the position on the horizontal axis and the value of the other variable determining the position on the vertical axis.

Here we use the `scatter()` function. 

```python
plt.scatter(df["Random Float"], df["Random Integer"])

# Add labels and titles
plt.xlabel('Random Float')
plt.ylabel('Random Integer')
plt.title('Scatter Plot Example')

# Show the plot
plt.show()
```

Adding a regression line to a scatter plot allows to inspect the relationship between two variables of a dataset.

```python
# Scatter plot
plt.scatter(df["Random Float"], df["Random Integer"])

# Calculate regression line
coefficients = np.polyfit(df["Random Float"], df["Random Integer"], 1)
regression_line = np.poly1d(coefficients)

# Plot regression line
plt.plot(df["Random Float"], regression_line(df["Random Float"]), color='red')

# Add labels and title
plt.xlabel('Random Float')
plt.ylabel('Random Integer')
plt.title('Scatter Plot with Regression Line')

# Show the plot
plt.show()
```

**Box plot**

A box plot, also known as a box-and-whisker plot, is a graphical representation of a distribution of a dataset. It is a way to visualize the five-number summary of a dataset, which consists of the minimum, first quartile, median, third quartile, and maximum.
The box plot uses a rectangle, called the box, to represent the interquartile range (the middle 50% of the data) and a line inside the box to represent the median. The "whiskers" outside of the box represent the minimum and maximum values of the data, while any data points outside the whiskers can be considered outliers.
The primary purpose of a box plot is to give a clear picture of the distribution of the data, including any outliers. Box plots are particularly useful for comparing the distribution of multiple datasets.

Here we use the `boxplot()` function. 

```python
df.columns.tolist()
```

```python
plt.boxplot(df)
plt.xlabel('Datasets')
plt.ylabel('Values')
plt.title('Box plot Example')
plt.xticks([1, 2], df.columns.tolist())
plt.show()
```
