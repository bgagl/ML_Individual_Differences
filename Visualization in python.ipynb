**Histogram**

A histogram is a graphical representation of the distribution of a continuous variable. The graph shows the frequency of different numerical ranges. The x-axis represents the values of the variable, and the y-axis represents the frequency of those values. The graph is divided into several "bins". The height of each bar in the histogram corresponds to the number of observations in the bin.

Here we use the `hist()` function and add labels and title. Note, to select a variable in a dataframe with `df["variable name"]`.


```python
plt.hist(df["Random Float"])
plt.xlabel('Random Float')
plt.ylabel('Frequency')
plt.title('My Histogram')
plt.show()
```

**Scatter Plot**

A scatter plot displays values of typically two variables of a data set to observe relationships between variables. The data is displayed as a collection of points, each having the value of one variable determining the position on the horizontal axis and the value of the other variable determining the position on the vertical axis.

```python
plt.scatter(df["Random Float"], df["Random Integer"])

# Add labels and titles
plt.xlabel('Random Float')
plt.ylabel('Random Integer')
plt.title('Scatter Plot Example')

# Show the plot
plt.show()
```

**Box plot**

A box plot, also known as a box-and-whisker plot, is a graphical representation of a distribution of a dataset. It is a way to visualize the five-number summary of a dataset, which consists of the minimum, first quartile, median, third quartile, and maximum.
The box plot uses a rectangle, called the box, to represent the interquartile range (the middle 50% of the data) and a line inside the box to represent the median. The "whiskers" outside of the box represent the minimum and maximum values of the data, while any data points outside the whiskers can be considered outliers.
The primary purpose of a box plot is to give a clear picture of the distribution of the data, including any outliers. Box plots are particularly useful for comparing the distribution of multiple datasets.

```python
plt.boxplot(df)
plt.xlabel('Datasets')
plt.ylabel('Values')
plt.title('Box plot Example')
plt.xticks([1, 2], df.columns.tolist())
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some data
np.random.seed(0)
x = np.linspace(0, 10, 10)
y = x + np.random.normal(0, 1, 10)
yerr = np.random.normal(2, 1, 10)

# Plot the data with error bars
plt.errorbar(x, y, yerr=yerr, fmt='o', color='black', ecolor='lightgray', elinewidth=3, capsize=0)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Generate some data
np.random.seed(0)
x = np.random.rand(100, 1)
y = 2 + 3 * x + np.random.rand(100, 1)

# Create a LinearRegression object
model = LinearRegression()

# Fit the model to the data
model.fit(x, y)

# Make predictions using the model
y_pred = model.predict(x)

# Plot the data and the model's predictions
plt.scatter(x, y, label='Data')
plt.plot(x, y_pred, color='r', label='Regression model prediction')
plt.legend()
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Scatter plot with linear regression line Example')
# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import numpy as np

# Generate some example data
x = np.linspace(0, 10, 50)
y = x**2 + np.random.normal(0, 4, 50)

# Define the function to fit (in this case, a polynomial function)
def func(x, a, b, c):
    return a*x**2 + b*x + c

# Fit the function to the data
popt, _ = curve_fit(func, x, y)

# Create the scatter plot
plt.scatter(x, y, label='data')

# Plot the fitted function
plt.plot(x, func(x, *popt), 'r-', label='fit: a=%5.3f, b=%5.3f, c=%5.3f' % tuple(popt))
plt.legend()
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Scatter plot with non-linear regression line Example')
# Show the plot
plt.show()
```
