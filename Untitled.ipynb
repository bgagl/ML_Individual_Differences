# Using categorization methods for predicting individual categories

Leave One Out Cross-validation (LOOCV) is a technique for assessing the performance of a machine learning model by training it on all but one of the available data points and evaluating it on the single left-out data point. This process is repeated for each data point, so that each point in the dataset is used once as the test set.

The scores for each iteration are collected in the "scores" list and the mean score is calculated and printed at the end.

Note that LOOCV can be computationally expensive, as it requires training the model as many times as there are data points. In practice, other techniques such as K-Fold cross-validation may be more computationally efficient.

```python
from sklearn.model_selection import LeaveOneOut
from sklearn.linear_model import LogisticRegression
from pandas import read_csv, melt
df = read_csv("./datasets/Group_A_B.csv")
df_long = melt(df, id_vars=None, value_vars=['NoProblem', 'Problem'], var_name='Group', value_name='Measure')

X = df_long["Measure"].values.reshape(-1, 1)
y = df_long["Group"]
```

```python
loo = LeaveOneOut()
model = LogisticRegression()

scores = []
for train_index, test_index in loo.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    model.fit(X_train, y_train)
    scores.append(model.score(X_test, y_test))

print("Scores: ", scores)
print("Mean score: ", sum(scores) / len(scores))
```
